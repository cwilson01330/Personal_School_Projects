## Part II: Model Justification

### B.

### 1.

Random forests are in the simplest terms a group of decision trees. This leads to a variety of positives including reduced chances of overfitting and normally an increase in the models ability to predict more correctly the test data set. Random forests take small samples of our training set to train individual trees, these trees make up a forest, the data for each tree being a small portion of the original data, but picked at random is where the random component comes from. The actual predicition is then made based on the majority count of trees in the voting pool. For the data at hand if 200 trees vote a patient will be readmitted and 30 vote they won't, then the patient will be predicted as a readmit. (z_ai, 2021)

### 2.

With random forests the only assumption is that the data sampling for each tree be representative of the whole data set. (Chatterjee, 2019)

## Part III. Data Preparation

### 1.

Like most models, Random forests don't have the ability to handle categorical variables in their native form, so to correct this, factorization must occur. Simply put factorizing categorical variables takes strings and converts them into integers. For example, Marital will convert Single, Married, Divorced, and Widowed into corresponding numbers such as 0,1,2,3 respectively. This process allows the model to use the data.

### 3.

The process of prepping this data for analysis is fairly simply due to the nature of it being so clean. First, columns that aren't crucial to the analysis are dropped. Then in the next code block, the categorical variables are factorized. Finally, the data is split into 2 variables, X contains the independent variables and y is the dependent.

```python
clean_data = data.drop(columns=['Area','CaseOrder','Customer_id','Interaction','UID','City','State','County','Zip','Lat','Lng','Population','TimeZone','Job','item1','item2','item3','item4','item5','item6','item7','item8'])
```

```python
clean_data
```

```python
cat_columns = clean_data.select_dtypes(['object']).columns
clean_data[cat_columns] = clean_data[cat_columns].apply(lambda x: pd.factorize(x)[0])
```

```python
clean_data.to_csv('clean_data.csv')
```

```python
X = clean_data.drop(columns=['ReAdmis'])
y = clean_data['ReAdmis']
```

## Part IV: Analysis

### D.

### 1.

```python
X_train,X_test,y_train,y_test = train_test_split(X, y,test_size=0.25,random_state=1,stratify=y)
```

### 2.

The process of analysis for a random forest is simply variable rf is created as an instance of the RandomForestClassifier, then the model is fitted with the training from the train_test_split. Y_pred is created as the predicitions for the dependent variable from the test X values and accuracy scores and Mean Squared Errors are calculated using sklearn metrics.

### 3.

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train,y_train);
```

```python
y_pred = rf.predict(X_test)
```

```python
rf.score(X_test, y_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
cv_scores = cross_val_score(rf, X, y, cv=5)
```

```python
cv_scores.mean()
```

## Part V: Data Summary and Implications

### E.

### 1.

```python
print('The accuracy score of the model is  %s' % (rf.score(X_test,y_test)*100))
```

The accuracy score of a model can be defined as the model's ability to correctly predict values. The model uses it's fitting to predict values for y_test based on the variables from X_test, those predictions are then checked against y_test's actual values and a score is generated. This model achieved and accuracy score of .934 or 93.4%. (Classification: Accuracy | Machine Learning Crash Course)

```python
print('The mean squared error of the model is  %s' % mean_squared_error(y_test,y_pred))
```

The mean squared error of a model can be defined as, as the name implies a square of the errors. This is most easily graphed, then the distance between the predicition line and the estimated values are squared and the resulting values are added together. SkLearn handles this calculation easily and takes all the heavy lifting and chance for human error. ( Binieli, 2018)
